```python
import matplotlib.pyplot as plt

a = int(input("Enter the starting number of your range : "))
b = int(input("Enter the ending number of your range : "))
numbers = []

print("Choose the filtering logic")
print("1 - All numbers in a range")
print("2 - Multiple of a given number")
print("3 - Prime numbers")
choice =int(input("Enter your choice(1/2/3) : "))

if choice == 1 :
    numbers = list(range(a, b+1))

elif choice == 2:
    print("Note: Multiples of an even number will all be even.")
    c = int(input("Enter the multiple to filter (e.g., 3 for multiples of 3): "))
    numbers = [x for x in range(a, b+1) if x % c == 0]

elif choice == 3 :
    def is_prime(n):
        if n < 2 :
            return False
        for j in range(2, int(n**0.5)+1):  
            if n % j == 0:
                return False
        return True
    numbers = [y for y in range(a, b+1) if is_prime(y)]

even_counter = []
odd_counter = []
for i in numbers:
    if(i%2 == 0):
        even_counter.append(i)
    else:
        odd_counter.append(i)

even_number_freq = len(even_counter)
odd_number_freq = len(odd_counter)

print(f"Total numbers checked = {len(numbers)}")
print(f"Total no. of even numbers in the given range = {even_number_freq}")
print(f"Total no. of odd numbers in the given range = {odd_number_freq}")

sizes = [even_number_freq, odd_number_freq]
label  = ["Even", "Odd"]
colors = ["skyblue","pink"]

plt.pie(sizes, labels=label , colors=colors ,autopct= '%1.1f%%', startangle= 120)
plt.title("Even and Odd numbers distribution")
plt.show()


```
